

*  Построим ***дерево отрезков*** - **O(2n)**.
*  В *дереве отрезков* будем хранить индексы текущего разбиения и текущую высоту. На картинке снизу приведен пример сохранения текущего разбиения на каждом уровне.
*  Поскольку функция f работает на любом под отрезке **O(1)**, то мы можем найти результат для каждого элемента из *дерева отрезков* за **O(2n)**.
*  После этого будем подниматься снизу вверх по *дереву отрезков* и передавать вверх наибольший результат из суммы разбиения на текущем уровне и значением в родителе, также фиксируя высоту с которой берется максимальное разбиение - **O(2n)**.
*  Так мы найдем максимальное разбиение на не пересекающие отрезки.
*  Для восстановления разбиения пройдемся по дереву сверху вниз и ищем переданный высоты из каждого поддерева - **O(2n)**.
*  Итоговая асимптотика: **O(2n)**

```python
from transformers import pipeline

distilled_student_sentiment_classifier = pipeline(
    model="lxyuan/distilbert-base-multilingual-cased-sentiments-student",
    return_all_scores=True
)
```

```python
# english
print(distilled_student_sentiment_classifier ("I love this movie and i would watch it again and again!"))

# malay
print(distilled_student_sentiment_classifier("Saya suka filem ini dan saya akan menontonnya lagi dan lagi!"))

# japanese
print(distilled_student_sentiment_classifier("私はこの映画が大好きで、何度も見ます！"))

# russian
print(distilled_student_sentiment_classifier("Это пиво просто отвратительно на вкус"))
```
